## 1. Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KernelDensity

# Display full output 
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_colwidth', None)
```

## 2. Importing Data

```python
spray = pd.read_pickle('./data/spray.pk1')
weather = pd.read_pickle('./data/weather.pk1')
train = pd.read_pickle('./data/train.pk1')
test = pd.read_pickle('./data/test.pk1')
```

## 4. Exploratory Data Analysis

### 4.1 Train Data

```python
# Plot heatmap using seaborn to visualize correlations between wnvpresent vs all numerical features

# Set the plot figure size
plt.figure(figsize = (22,20))

# Find the pairwise correlation of all columns in dataframe
corr = combine_train.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr)        
mask[np.triu_indices_from(mask)] = True

# Draw the heatmap
sns.heatmap(corr, mask = mask, cmap = 'coolwarm',  annot = True)

# Set the heatmap title
plt.title('Correlations of West Nile Virus vs numerical features', fontsize = 14);  
```

```python
# Plot heatmap using seaborn to visualize correlations between wnvpresent vs all numerical features

# Set the plot figure size
plt.figure(figsize = (10,15))

# Find the pairwise correlation of all columns in dataframe
corr = combine_train.corr()

# Draw the heatmap
sns.heatmap(corr[['wnvpresent']].sort_values(by = 'wnvpresent', ascending = False), cmap = 'coolwarm',  annot = True)

# Set the heatmap title
plt.title('Correlations of West Nile Virus vs numerical features', fontsize = 14);
```

```python
alpha_cm = plt.cm.Greens
alpha_cm._init()
alpha_cm._lut[:-3,-1] = abs(np.logspace(0, 1, alpha_cm.N) / 10 - 1)[::-1]

# plot locations of traps and instances of WNV cases
mapdata = np.loadtxt("./assets/mapdata_copyright_openstreetmap_contributors.txt")
aspect = mapdata.shape[0] * 1.0 / mapdata.shape[1]
lon_lat_box = (-88, -87.5, 41.6, 42.1)

# Spray location
X = spray[['longitude', 'latitude']].drop_duplicates().values
kd = KernelDensity(bandwidth=0.015)
kd.fit(X)

xv,yv = np.meshgrid(np.linspace(-88, -87.5, 100), np.linspace(41.6, 42.1, 100))
gridpoints = np.array([xv.ravel(),yv.ravel()]).T
zv = np.exp(kd.score_samples(gridpoints).reshape(100,100))

#kernel density
plt.figure(figsize=(10,14))
plt.imshow(mapdata, cmap = plt.get_cmap('gray'), extent = lon_lat_box, aspect = aspect)
plt.imshow(zv, origin = 'lower', cmap = alpha_cm, extent = lon_lat_box, aspect = aspect)

# trap locations
trap_locations = combine_train[['longitude', 'latitude']].values
plt.scatter(trap_locations[:,0], trap_locations[:,1], marker = 'o', c = 'b', label = 'Trap Locations')

# wnn outbreak locations
wnv_locations = combine_train[combine_train['wnvpresent'] != 0][['longitude', 'latitude']].values
plt.scatter(wnv_locations[:,0], wnv_locations[:,1], marker = '*', c = 'r', label = 'WNV Outbreak Locations')

plt.title('Spray Locations in Chicago', fontsize = 18)
plt.legend(fontsize = 12)
plt.xlabel('Longitude', fontsize = 12)
plt.ylabel('Latitude', fontsize = 12)
plt.savefig('./image/Spray Locations.png')
```

<b>Observation<b/> : 

1. The blue circles indicate the location of the traps and the red stars indicate the location of the West Nile Virus outbreak.
2. The green cloud indicate the rough coverage areas that were sprayed. Area with the darker green indicates that the region has higher spray concentration/frequency and the area with lighter green means that less spray concentration/frequency.
3. All the traps are spread out quite evenly, however the spray system fails to cover all the locations with the West Nile Virus outbreak.

```python
alpha_cm1 = plt.cm.Reds
alpha_cm1._init()
alpha_cm1._lut[:-3,-1] = abs(np.logspace(0, 1, alpha_cm1.N) / 10 - 1)[::-1]

# plot locations of traps and instances of WNV cases
mapdata = np.loadtxt("./assets/mapdata_copyright_openstreetmap_contributors.txt")
aspect = mapdata.shape[0] * 1.0 / mapdata.shape[1]
lon_lat_box = (-88, -87.5, 41.6, 42.1)

#Traps where only WNV is present
wnv = combine_train[combine_train['wnvpresent'] == 1]
wnv = wnv.groupby(['date', 'trap','longitude', 'latitude']).max()['wnvpresent'].reset_index()
X1 = wnv[['longitude', 'latitude']].values
kd1 = KernelDensity(bandwidth=0.015)
kd1.fit(X1)

xv,yv = np.meshgrid(np.linspace(-88, -87.5, 100), np.linspace(41.6, 42.1, 100))
gridpoints = np.array([xv.ravel(),yv.ravel()]).T
zv1 = np.exp(kd1.score_samples(gridpoints).reshape(100,100))

#kernel density
plt.figure(figsize = (10,14))
plt.imshow(mapdata, cmap = plt.get_cmap('gray'), extent = lon_lat_box, aspect=aspect)
plt.imshow(zv1, origin = 'lower', cmap=alpha_cm1, extent = lon_lat_box, aspect=aspect)

# trap locations
trap_locations = combine_train[['longitude', 'latitude']].drop_duplicates().values
plt.scatter(trap_locations[:,0], trap_locations[:,1], marker='o', c = 'b', label='Trap Locations')

# wnn outbreak locations
wnv_locations = combine_train[combine_train['wnvpresent'] != 0][['longitude', 'latitude']].values
plt.scatter(wnv_locations[:,0], wnv_locations[:,1], marker='*', c = 'r', label = 'WNV Outbreak Locations')

plt.title('West Nile Virus Outbreak Locations in Chicago', fontsize = 18)
plt.legend(fontsize = 12)
plt.xlabel('Longitude', fontsize = 12)
plt.ylabel('Latitude', fontsize = 12)
plt.savefig('../image/WNV Outbreak Locations.png')
```

<b>Observation<b/> : 

1. Area with the darker red indicates that the region has a high number mosquitoes that carry the West Nile Virus. Area with lighter red means that the region has a low number mosquitoes that carry the West Nile Virus.

```python
#Number of mosquitos per year and per species
year_mos = combine_train.groupby(['year','species'], as_index=False).agg({'nummosquitos' : 'sum' }).reset_index()

plt.figure(figsize = (16,7))
sns.barplot(data = year_mos, x = 'year', y = 'nummosquitos', hue = 'species')
plt.title('Number of mosquitoes of each species over the years',fontsize = 14)
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Number of mosquitoes', fontsize = 14);
```

```python
wnv_species = combine_train[combine_train['wnvpresent'] == 1]
```

```python
wnv_species['species'].value_counts()
```

```python
plt.figure(figsize = (8,5))
sns.countplot(data = wnv_species, x = 'species')
plt.title('Mosquitoes Species with West Nile Virus', fontsize = 12)
plt.xlabel('Mosquitoes Species', fontsize = 12)
plt.ylabel('Count', fontsize = 12);
```

<b>Observation<b/> : 

1. From the graph above, we can see that Culex Pipiends and Culex Restuans are the most captured mosquitos and the only <b>species</b> that will carry West Nile Virus.

```python
plt.figure(figsize = (10,6))
sns.barplot(data = combine_train.groupby('month').sum().reset_index(), x = 'month', y = 'wnvpresent', color = 'springgreen')
plt.title('West Nile Virus Present Months (2007, 2009, 2011, 2013)', fontsize = 12)
plt.xlabel('Month', fontsize = 12)
plt.ylabel('West Nile Virus Present', fontsize = 12);
```

<b>Observation<b/> : 

1. From the graph above, we can see that West Nile Virus occur/spread started from July till September with August as peak, which is the summer season in Chicago. This is proof that warm and moist weather is mosquito's friend.

```python
plt.figure(figsize = (10,6))
sns.lineplot(data = combine_train, x = 'month', y = 'wnvpresent', hue = 'year', ci = None, palette = 'coolwarm')
plt.legend(fontsize = 12)
plt.title('Monthly West Nile Virus occurrences by year', fontsize = 12)
plt.xlabel('Month', fontsize = 12)
plt.ylabel('West Nile Virus occurrences', fontsize = 12)

plt.tight_layout()
```

<b>Observation<b/> : 

1. Dew Point - The dew point is the temperature to which air must be cooled to become saturated with water vapor. So if the air is cooling at a higher temp that means it has already enough of moisture present and it just needs to be cooled by a very minimal amount before it starts condensing into water vapour.
2. Wet Bulb - Similar is the case with Wet Bulb. If the Wet Bulb temp is significantly lower then it means the air earlier was pretty dry and vice-versa.
3. From the line graph above, we can see that August is the peak moment for the West Nile Virus occurrences every year and 2013 has the highest occurences of West Nile Virus occurrences present.

```python
plt.figure(figsize = (8,4))
sns.lmplot(data = combine_train, x = 'wetbulb', y = 'dewpoint', fit_reg = False, hue=  'wnvpresent', height = 7, aspect = 0.9)
plt.title('Mosquitoes with West Nile Virus by Wet Bulb Temperature(F) & Dew Point Temperature(F)', fontsize = 12)
plt.xlabel('Wet Bulb Temperature(F)', fontsize = 12)
plt.ylabel('Dew Point Temperature(F)', fontsize = 12);
```

<b>Observation<b/> : 

1. Since mosquitoes love warm and moist weather we can see that most of the West Nile Virus occurrences happened when the days with the wet bulb temperature above 56 degrees Fahrenheit.
